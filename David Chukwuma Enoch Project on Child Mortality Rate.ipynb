```python
# Filtering rows for global data
global_data = df7[df7['Entity'] == 'World']

# Selecting vaccination columns
vaccination_columns = [col for col in df7.columns if '(%' in col]

# Calculating global averages by year for all vaccines
global_averages = global_data[['Year'] + vaccination_columns].set_index('Year')

print(global_averages)
```

```python
# Line plot: Vaccination trends over years
plt.figure(figsize=(10, 6))
for col in vaccination_columns:
    plt.plot(global_averages.index, global_averages[col], label=col)

plt.title('Global Vaccination Coverage Trends')
plt.xlabel('Year')
plt.ylabel('Vaccination Coverage (%)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Pivoting the data to create a 2D structure for the heatmap
heatmap_data = global_averages.T  # Transpose for years as columns and vaccines as rows

# Create heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap="coolwarm")

# Add titles and labels
plt.title('Global Vaccination Coverage Trends')
plt.xlabel('Year')
plt.ylabel('Vaccine')
plt.show()
```
